```python
import wikipedia
import numpy as np
import os
import json
import requests
```

```python
code2lang = wikipedia.languages()
langs = ['ru', 'uk', 'en', 'nl', 'de', 'bg']
```

```python
def get_texts_for_lang(lang, n=100): # функция для скачивания статей из википедии
    wikipedia.set_lang(lang)
    wiki_content = []
    pages = wikipedia.random(n)
    
    for page_name in pages:
        try:
            page = wikipedia.page(page_name)
        
        except Exception as e:
            print('Skipping page {}'.format(page_name), e)
            continue

        wiki_content.append('{}\n{}'.format(page.title, page.content.replace('==', '')))

    return wiki_content
```

```python
wiki_texts = {}
for lang in langs:
    try:
        wiki_texts[lang] = get_texts_for_lang(lang, 100)
    except Exception as e:
        print('ERROR ON - ', lang, e)
        continue
    print(lang, len(wiki_texts[lang]))
```

Частотные списки

```python
import codecs
import collections
import sys
import re
import operator

def tokenize(text):
    text = text.lower()
    text = re.sub(r'\s+', ' ', text, re.UNICODE)
    text = re.sub(r'[^\w\s]', '', text, re.UNICODE)
    return text.split(' ')


def freq(wiki_texts, lang):
    freqs = collections.defaultdict(lambda: 0)
    words = []
    corpus = wiki_texts[lang]
    for article in corpus:
        for word in tokenize(article.replace('\n', '').lower()):
            freqs[word] += 1

    for word in sorted(freqs, key=lambda w: freqs[w], reverse=True)[:50]:
        words.append(word)
#         print('{}\t{}'.format(freqs[word], word))
    return words
```

```python
freq_ = {}
for lang in langs:
    freq_[lang] = freq(wiki_texts, lang)
```

```python
def coun(freq_, text):
    text = text.lower()
    i_ = {}
    i = 0
    for key, item in freq_.items():
        for word in item:
#         print(item)
            if word in text.split():
                i += 1
        i_[key] = i
        i = 0
    print(i_)
    return max(i_.items(), key=operator.itemgetter(1))[0]
```

```python
text = wiki_texts
print(text)
cc = coun(freq_, text)
cc
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
true_labels = []
predicted_labels = []

for lang in wiki_texts:
    for text in wiki_texts[lang]:
        true_labels.append(lang)
        predicted_labels.append(coun(freq_,text))
```

```python
print(classification_report(true_labels, predicted_labels))
```

Набор N-gramm

```python
from itertools import islice, tee

def make_ngrams(text):
    N = 3 # задаем длину n-граммы
    ngrams = zip(*(islice(seq, index, None) for index, seq in enumerate(tee(text, N))))
    ngrams = [''.join(x) for x in ngrams]
    return ngrams
```

```python
def freq_ns(wiki_texts, lang):
    freqs = collections.defaultdict(lambda: 0)
    ns = []
    corpus = wiki_texts[lang]
    for article in corpus:
        for ngram in make_ngrams(article.replace('\n', '').lower()):
            freqs[ngram] += 1

    for ngram in sorted(freqs, key=lambda n: freqs[n], reverse=True)[:50]:
        ns.append(ngram)
#         print('{}\t{}'.format(freqs[word], word))
    return ns
```

```python
freq_n = {}
for lang in langs:
    freq_n[lang] = freq_ns(wiki_texts, lang)
```

```python
def coun_ns(freq_n, text):
    text = make_ngrams(text)
    i_ = {}
    i = 0
    for key, item in freq_n.items():
        for ns in item:
#         print(item)
            if ns in text:
                i += 1
        i_[key] = i
        i = 0
    print(i_)
    return max(i_.items(), key=operator.itemgetter(1))[0]
```

```python
text = wiki_texts['en'][0]
print(text)
nn = coun_ns(freq_n, text)
nn
```

```python
true_labels = []
predicted_labels = []

for lang in wiki_texts:
    for text in wiki_texts[lang]:
        true_labels.append(lang)
        predicted_labels.append(coun_ns(freq_n,text))
```
