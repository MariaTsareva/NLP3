```python
import pycrfsuite
import codecs
```

```python
def read_ruscorpora(filename):
    sent = [[]]
    with codecs.open(filename, encoding='utf-8') as inp_file:
        for line in inp_file:
            line = line.strip('\r\n ')
            if '\t' in line:
                line = line[:line.find('\t')]
            if line.startswith('#'):
                continue
            if not line:
                sent.append([])
                continue
            wordform, lemma, gram = line.rsplit('/', 2)
            pos = gram[:gram.find('=')] if '=' in gram else gram
            pos = pos.split(',')[0]
            sent[-1].append((wordform, pos, lemma, gram))
    
    return sent
```

```python
import numpy as np

data_fr = pd.DataFrame()
Word = []
left = []
right = []
lemma = []
pos_ = []
for sent in corpus:
    for i, word in enumerate(sent):
        lemma.append(word[2])
        pos_.append(word[1])
        Word.append(word[0])
        if i == 0:
            left.append(np.NaN)
        else:
            left.append(sent[i-1][1])
        
        if i == len(sent)-1:
            right.append(np.NaN)
        else:
            right.append(sent[i+1][1])

data_fr['Word'] = Word
data_fr['Lemma'] = lemma
data_fr['POS'] = pos_
data_fr['left'] = left
data_fr['right'] = right
```
